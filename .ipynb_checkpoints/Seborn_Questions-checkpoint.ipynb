```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
#Plot a boxplot of fare across different classes (pclass).
sns.boxplot(x = 'pclass', y = 'fare', data = df, palette = 'Pastel1')

plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```

```python
# Plot a violin plot comparing age distributions by gender

plt.figure(figsize=(8,5))
sns.violinplot(x = 'sex', y = 'age', data = df, palette = 'Set2')

plt.title('Age Distribution by Gender')
plt.show()
```

```python
#Create a countplot showing passenger count by embarkation port (embarked).

sns.countplot(x = 'embarked', data = df, palette = 'Set3')
plt.title('Passenger Count by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
#Plot a pairplot of numeric features (age, fare, sibsp, parch).

cols = ['age', 'fare', 'sibsp', 'parch']

sns.pairplot(df[cols].dropna())   # dropna to avoid errors with missing values
plt.suptitle('Pairplot of Numeric Features', y=1.02)
plt.show()    
```

```python
#Heatmap of correlations among numeric columns

numeric_df = df.select_dtypes(include='number')

plt.figure(figsize=(8,6))

sns.heatmap(numeric_df.corr(), annot = True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Titanic Dataset')
plt.show()
```

```python
#Barplot showing average fare by pclass

plt.figure(figsize=(8,5))
sns.barplot(x = 'pclass', y = 'fare', data = df, palette='Blues_d')
plt.title('Average Fare by Passenger Class')
plt.show()
```

```python
#Scatterplot of fare vs age with hue by survived

plt.figure(figsize=(8,5))
sns.scatterplot(x = 'age', y = 'fare', data = df, hue = 'survived', palette='Set1')
plt.title('Fare vs Age (Survived vs Not Survived)')
plt.show()
```

```python
#Swarmplot to visualize fare distribution by pclass

plt.figure(figsize=(8,5))
sns.swarmplot(x='pclass', y='fare', data=df, palette='Set2')
plt.title('Fare Distribution by Class')
plt.show()
```

```python
#Lineplot showing average fare vs age grouped by sex

sns.lineplot(x = 'age', y = 'fare', hue = 'sex', data = df.dropna(), estimator = 'mean')
plt.title('Average Fare vs Age by Gender')
plt.show()
```

```python
#FacetGrid to visualize age distribution across gender and class

g = sns.FacetGrid(df, col='sex', row='pclass', margin_titles=True)
g.map(sns.histplot, 'age', bins=15, color='skyblue')
g.fig.suptitle('Age Distribution by Gender and Class', y=1.05)
plt.show()
```
