```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Q1. Create a dataset with the following columns-> Products,Regions,Sales,Profit and 10 rows:
data = {
    'Products': ['Laptop', 'Smartphone', 'Headphones', 'Monitor', 'Keyboard',
                 'Mouse', 'Tablet', 'Webcam', 'Speaker', 'External Drive'],
    'Regions': ['North', 'South', 'East', 'West', 'North',
                'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 5000, 10),  # Random sales figures
    'Profit': np.random.randint(-50, 800, 10)   # Random profit figures (some losses)
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
#Q2. Plot a line chart of Sales for all 10 rows (index as X-axis).

plt.figure(figsize= (10, 5))
sns.lineplot(y = df['Sales'], x = df['Products'])
```

```python
df['Sales'].plot(
    kind='line',
    marker='o',         # Use circles for data points
    linestyle='-',      # Use a solid line
    color='teal'        # Line color
)
```

```python
#Q3. Plot a bar chart showing total sales for each product.

sns.barplot(x = df['Products'], y = df['Sales'], estimator= np.sum)
plt.xticks(rotation=45, ha='right')
```

```python
#Q4. Plot a scatter plot of Sales vs Profit.

sns.scatterplot(x = df['Sales'], y = df['Profit'])
```

```python
#Q5. Create a pie chart showing total sales by region.

region_sales = df.groupby('Regions')['Sales'].sum()

plt.pie(
    region_sales.values, 
    labels=region_sales.index,
    autopct='%1.1f%%',   # Display percentage with one decimal place
    startangle=90,       # Start the plot from the top
    shadow=True,
    explode=(0.05, 0, 0, 0) 
)
```

```python
#Q6. Plot a histogram of the Sales column.

sns.histplot(df['Sales'], bins = 20, kde = True)
```

```python
#Q7. Create a boxplot of Profit to detect outliers.

sns.boxplot(x = 'Profit', data = df)
```

```python
#Q8. Create a countplot showing how many products are sold in each region.

sns.countplot(x = df['Regions'])
```

```python
#Q9. Create a heatmap to show correlation between Sales and Profit.

correlation_matrix = df[['Sales', 'Profit']].corr()

sns.heatmap(correlation_matrix, annot = True, cmap='coolwarm',
    fmt=".2f",
    linewidths=0.5,
    linecolor='black'
)
```

```python
#Q10. Show total and average sales per region using pandas.

region_summary = df.groupby('Regions')['Sales'].agg(['sum', 'mean'])
region_summary
```
